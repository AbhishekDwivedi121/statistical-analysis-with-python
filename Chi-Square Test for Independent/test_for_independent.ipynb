# Chi-Square Test for Independence

The **Chi-Square Test for Independence** is a statistical test used to determine whether there is a **significant association** between two **categorical variables**.

---

## ðŸ”¹ Definition

The Chi-Square Test for Independence helps us check whether two variables are **independent** or **related** to each other.

It compares the **observed frequencies** (actual data) with the **expected frequencies** (data we would expect if the variables were independent).

---

## ðŸ”¹ Formula

The formula for the Chi-Square statistic is:

$$
\chi^2 = \sum \frac{(O - E)^2}{E}
$$

Where:

- $O$ = Observed frequency  
- $E$ = Expected frequency  
- $\sum$ = Summation over all possible categories  

---

## ðŸ”¹ Expected Frequency Formula

Expected frequency for each cell is calculated as:

$$
E = \frac{(\text{Row Total}) \times (\text{Column Total})}{\text{Grand Total}}
$$

---

## ðŸ”¹ Hypotheses

- **Null Hypothesis ($H_0$):** The two variables are independent.  
- **Alternative Hypothesis ($H_1$):** The two variables are not independent (there is an association).

---

## ðŸ”¹ Degrees of Freedom (df)

The degrees of freedom for the test are given by:

$$
df = (r - 1)(c - 1)
$$

Where:
- $r$ = Number of rows  
- $c$ = Number of columns  

---

## ðŸ”¹ Decision Rule

- If the calculated value of $\chi^2$ > critical value from Chi-Square table â†’ **Reject $H_0$**  
- Otherwise â†’ **Fail to reject $H_0$**

---

## ðŸ”¹ Example (Conceptual)

Suppose we want to test whether **gender** and **preference for a product** are related.  
We collect data and create a contingency table, then apply the Chi-Square formula to test independence.

---

### Example â€” Chi-Square Test for Independence

We are testing whether **Smoker** and **Gender** are independent in the **Tips dataset**.

- **Null Hypothesis ($H_0$):** Smoker and Gender are independent.  
- **Alternative Hypothesis ($H_1$):** Smoker and Gender are not independent.

After applying the Chi-Square Test:

$$
\chi^2 = \sum \frac{(O - E)^2}{E}
$$

If the *p-value* is less than 0.05, we **reject $H_0$**, meaning there is a **significant relationship** between *smoking status* and *gender*.

Otherwise, we **fail to reject $H_0$**, meaning *smoking* and *gender* are **independent** in the Tips dataset.


```python
tabel = pd.crosstab( data['smoker'], data['sex'])
tabel
```

```python
def test_for_independence(observed):
    arr_tabel = np.array(observed)
    row_sum = arr_tabel.sum(axis=1)
    column_sum = arr_tabel.sum(axis=0)
    grand_sum = arr_tabel.sum()
    
    expected = np.zeros(arr_tabel.shape)
    
    for i in range(arr_tabel.shape[0]):
        for j in range(arr_tabel.shape[1]):
            expected[i, j] = (row_sum[i] * column_sum[j]) / grand_sum
            
    print("Observed-Table")
    print(arr_tabel)
    print('___________________________________')
    print("Expected-Table")
    print(expected)
    print('___________________________________')
    
    chi_square = ((arr_tabel - expected) ** 2) / expected
    chi_square = chi_square.sum()
    print("Chi-square value")
    print(chi_square)
    print('___________________________________')
    
    print("Degree-of-freedom")
    row = arr_tabel.shape[0]
    column = arr_tabel.shape[1]
    df = ((row - 1) * (column - 1))
    print(df)
    print('___________________________________')
    print("P-value")
    p_value = chi2.sf(chi_square, df)
    print(p_value)
    print('___________________________________')
    print("Conclusion")
    if p_value < 0.05:
        print("âœ… H1 Wins")   
    else:
        print("âŒ H0 Wins")       
```

```python
test_for_independence(tabel)
# Since the p-value is greater than 0.05 
# we fail to reject the null hypothesis 
# This means there is no significant relationship 
# between smoker and gender in the Tips dataset â€” they appear to be independent of each other.
```

### Example â€” Chi-Square Test for Independence (Titanic Dataset)

Now, we are working on the **Titanic dataset** to test whether **Survival status (Survived)** has any relationship with **Passenger Class (Pclass)**.

- **Null Hypothesis ($H_0$):** Survived and Passenger Class are independent.  
- **Alternative Hypothesis ($H_1$):** Survived and Passenger Class are not independent.

We will apply the Chi-Square Test for Independence to see if **passenger class** had any significant effect on **survival chances**.


```python
tabel = pd.crosstab(titanic['class'], titanic['survived'])
tabel
```

```python
test_for_independence(tabel)
# The Chi-Square test result shows that the p-value is less than 0.05 
# so we reject the null hypothesis 
# This means there is a significant relationship between 
# Survival status and Passenger Class in the Titanic dataset 
# passengers from higher classes were more likely to survive.
```
